**CREATING TOKEN EMBEDDINGS**

```python
import torch
input_ids = torch.tensor([2,3,5,1])
```

For the sake of simplicity and illustration purpose , suppose we have a small vocabulary of only 6 words (instead of 50257 words in the BPE tokenizer vocabulary), and we want to create embeddings of size 3 (in GPT-3, the embedding size is 12,288 dimensions).

```python
vocab_size = 6
output_dim = 3

torch.manual_seed(123) # For reproducability purpose.
embedding_layer = torch.nn.Embedding(vocab_size,output_dim)
print(embedding_layer.weight)
```

```python
print(embedding_layer(torch.tensor([3]))) # Access the embedding matrix.Same as 4th row as above. Works like an LUT.
```

```python
print(embedding_layer(input_ids)) # Multiple ID's.
```
